```python
import numpy as np
import scipy.stats as ss
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
num = 1000
feature = 500
```

```python
def HSIC(X,Y):
    return np.trace(kernel(X,func)@J@kernel(Y,func)@J)/(num-1)**2
def kernel(X,func):
    One = np.ones((num,1))
    phi = (np.linalg.norm(X,axis=0)**2).reshape(num,1)
    Phi1 = phi@One.T
    Phi2 = One@phi.T
    inner = X.T@X
    return func(Phi1+Phi2-2*inner)
J = np.eye(num)-np.ones((num,num))/num
func = lambda x: np.exp(-x/feature)
# func = lambda x: np.exp(-x)
```

```python
alpha = np.random.uniform(1/feature,1/feature,size=(feature,feature))

X = ss.norm.rvs(loc=0,scale=1,size=(feature,num))
N = ss.gamma.rvs(a=10,size=(feature,num))
Y = alpha@X + N
```

```python
plt.figure(figsize=(12.,8))
plt.scatter(X[:,0],X[:,1],alpha=0.5)
plt.scatter(Y[:,0],Y[:,1],color='r',alpha=0.5)
plt.show()
```

```python
beta_x2y = np.linalg.inv(X@X.T)@X@Y.T
beta_y2x = np.linalg.inv(Y@Y.T)@Y@X.T
HSIC(X,Y-beta_x2y@X)
```

```python
HSIC(Y,X-beta_y2x@Y)
```
